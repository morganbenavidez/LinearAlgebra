```python
class UnitsSiSystem:

  def __init__(self, number, prefix, base, convertToPrefix, convertToBase):

    self.number = number
    self.prefix = prefix
    self.base = base
    self.convertToPrefix = convertToPrefix
    self.convertToBase = convertToBase
    self.convert()

  def convert(self):

    # Base Units
    # (exponent in base 10, string representation)
    
    # Mass
    Gram = (0, 'gram')
    # Length
    Meter = (0, 'meter')
    # Time
    Second = (0, 'second')
    # Temperature
    Kelvin = (0, 'Kelvin')
    # Amount
    Mole = (0, 'mole')
    # Current
    Ampere = (0, 'ampere')
    # Intensity
    Candela = (0, 'candela')

    BaseDictionary = dict({'gram': Gram, 'meter': Meter, 'second': Second, 
                           'Kelvin': Kelvin, 'mole': Mole, 'ampere': Ampere, 'candela': Candela})

    # Prefixes and Definitions
    # (exponent in base 10, string representation)

    Giga = (.000000001, 'giga', 12)
    Mega = (.000001, 'mega', 11)
    Kilo = (.001, 'kilo', 10)
    Hecto = (.01, 'hecto', 9)
    Deka = (.1, 'deka', 8)
    Base = (1, '', 7)
    Deci = (10, 'deci', 6)
    Centi = (100, 'centi', 5)
    Milli = (1000, 'milli', 4)
    Micro = (1000000, 'micro', 3)
    Nano = (1000000000, 'nano', 2)
    Pico = (1000000000000, 'pico', 1)
    Femto = (1000000000000000, 'femto', 0)

    PrefixDictionary = dict({'giga': Giga, 'mega': Mega, 'kilo': Kilo, 'hecto': Hecto, 'deka': Deka, 
                             'base': Base, 'deci': Deci,'centi': Centi, 'milli': Milli, 'micro': Micro, 
                             'nano': Nano, 'pico': Pico, 'femto': Femto})


    try: 
  
      num = self.number
      cfPref = PrefixDictionary.get(self.prefix)
      cfBase = BaseDictionary.get(self.base)
      ctPref = PrefixDictionary.get(self.convertToPrefix)
      ctBase = BaseDictionary.get(self.convertToBase)

      answer = ((ctPref[0]/cfPref[0]) * num)

      return answer, ctPref[1] + ctBase[1]

    except KeyError as err:

      print('Acceptable Si Prefixes = giga, mega, kilo, hecto, deka, base, deci, centi, milli, micro, nano, pico, femto' + '\n' + 'Function accepts 5 parameters -> UnitsSiSystem(number, prefix, base, convertToPrefix, convertToBase)' + '\n' 
            + 'First Parameter(int) = value of starting measurement' + '\n' + 'Second Parameter(string) = Starting prefix of unit' + '\n' + 'Third Parameter(string) = Starting Base Unit' + '\n' + 'Fourth Parameter(string) = ending Prefix' + 
            '\n' + 'Fifth Parameter(string) = ending base unit') 
      raise

    except TypeError as err:

      print('Acceptable Si Prefixes = giga, mega, kilo, hecto, deka, base, deci, centi, milli, micro, nano, pico, femto' + '\n' + 'Function accepts 5 parameters -> UnitsSiSystem(number, prefix, base, convertToPrefix, convertToBase)' + '\n' 
            + 'First Parameter(int) = value of starting measurement' + '\n' + 'Second Parameter(string) = Starting prefix of unit' + '\n' + 'Third Parameter(string) = Starting Base Unit' + '\n' + 'Fourth Parameter(string) = ending Prefix' + 
            '\n' + 'Fifth Parameter(string) = ending base unit') 
      raise

    

```
