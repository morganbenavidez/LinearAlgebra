```python
class UnitsSiSystem:

  def __init__(self, number, prefix, base, convertToPrefix, convertToBase):

    self.number = number
    self.prefix = prefix
    self.base = base
    self.convertToPrefix = convertToPrefix
    self.convertToBase = convertToBase
    self.convert()

  def convert(self):

    # Base Units
    # (exponent in base 10, string representation)
    
    # Mass
    Gram = (0, 'gram')
    # Length
    Meter = (0, 'meter')
    # Time
    Second = (0, 'second')
    # Temperature
    Kelvin = (0, 'Kelvin')
    # Amount
    Mole = (0, 'mole')
    # Current
    Ampere = (0, 'ampere')
    # Intensity
    Candela = (0, 'candela')

    BaseDictionary = dict({'gram': Gram, 'meter': Meter, 'second': Second, 
                           'Kelvin': Kelvin, 'mole': Mole, 'ampere': Ampere, 'candela': Candela})

    # Prefixes and Definitions
    # (exponent in base 10, string representation)

    Giga = (9, 'giga')
    Mega = (6, 'mega')
    Kilo = (3, 'kilo')
    Root = (0, '')
    Deci = (-1, 'deci')
    Centi = (-2, 'centi')
    Milli = (-3, 'milli')
    Micro = (-6, 'micro')
    Nano = (-9, 'nano')
    Pico = (-12, 'pico')
    Femto = (-15, 'femto')

    PrefixList = [Femto, Pico, Nano, Micro, Milli, Centi, Deci, Kilo, Mega, Giga]

    PrefixDictionary = dict({'giga': Giga, 'mega': Mega, 'kilo': Kilo, 'root': Root, 'deci': Deci,
                                 'centi': Centi, 'milli': Milli, 'micro': Micro, 'nano': Nano,
                                 'pico': Pico, 'femto': Femto})

    num = self.number
    cfPref = PrefixDictionary.get(self.prefix)
    cfBase = BaseDictionary.get(self.base)
    ctPref = PrefixDictionary.get(self.convertToPrefix)
    ctBase = BaseDictionary.get(self.convertToBase)

    if (cfPref[0] == 0):

      newExponent = cfPref[0] + ctPref[0]

    elif (cfPref[0] > 0):

      newExponent = cfPref[0] - ctPref[0]

    elif ((cfPref[0] < 0) and (ctPref[0] != 0)):

      newExponent = abs(cfPref[0]) + ctPref[0]

    elif ((cfPref[0] < 0) and (ctPref[0] == 0)):

      newExponent = cfPref[0]

    print('\n')
    print('Analysis')
    print('New Exponent: ' + str(newExponent))
    print('10 raised to new exp: ' + str(10**newExponent))
    #print(num*(1/(10**newExponent)))
    print('Div: ' + str(num*(1/(10**newExponent))))
    print('Mult: ' + str(num*(10**newExponent)))
    theExponential = 10**newExponent

    if (newExponent == 0):

      return num, cfBase[1]

    elif (newExponent > 0):
      print('ZeeMultiplication')
      num = num*(10**newExponent)

    elif ((newExponent < 0) and (ctPref[1] == '')):
      print('ZeeMult')
      num = num*(10**newExponent)

    elif (newExponent < 0 and (ctPref[1] != '')):
      print('ZeeDiv')
      w = 10**newExponent
      print('10 raised to new exp: ' + str(w))
      v = 1/w
      print(v)
      num = num/(10**newExponent)

    return num, ctPref[1] + ctBase[1]


    #return newExponent
    #return cfPref, cfBase, ctPref, ctBase


```

```python
class ByteConverter():

  def __init__(self, number, startEnding, convertToEnding):

    self.number = number
    self.startEnding = startEnding
    self.convertToEnding = convertToEnding
    self.convert()

  def convert(self):

    bit = 1
    byte = 8 * bit
    kb = 1024 * byte
    mb = 1024 * kb
    gb = 1024 * mb
    tb = 1024 * gb
    pb = 1024 * tb
    eb = 1024 * pb  # Exabyte
    zb = 1024 * eb  # Zettabyte
    yb = 1024 * zb  # Yottabyte

    # Everything in this dictionary is in terms of bits
    conversionDictionary = dict({'bits': bit, 'bytes': byte, 'kb': kb, 'mb': mb, 'gb': gb, 'tb': tb, 'pb': pb, 'eb': eb, 'zb': zb, 'yb': yb})
    
    try: 
  
      startEndingValue = conversionDictionary[self.startEnding]

      convertToEndingValue = conversionDictionary[self.convertToEnding]

      fullyConverted = (startEndingValue / convertToEndingValue) * self.number

      return fullyConverted, self.convertToEnding

    except KeyError as err:

      print('Acceptable Units of Memory = bits, bytes, kb (kilobyte), gb (gigabyte), tb (terabyte), pb (petabyte), eb (exabyte), zb (zettabyte), yb (yottabyte)' + '\n' + 'Function accepts 3 parameters -> convert(int, string1, string2)' + '\n' 
            + 'First Parameter(int) = number of items trying to convert' + '\n' + 'Second Parameter(string1) = Starting Unit of Memory' + '\n' + 'Third Parameter(string2 = Unit of Memory you want to convert to') 
      raise
```
