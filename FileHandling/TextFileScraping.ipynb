```python
import os, glob


def openFiles(path):
   
    fileNameList = []
   
    #for filename in glob.glob(os.path.join(path, 'SpecificFileName.txt')): # Change this to bottom line to do all.
    for filename in glob.glob(os.path.join(path, '*.txt')):
        with open(os.path.join(os.getcwd(), filename), 'r') as f:
            
            tempList = []

            # Remove the path and leave just the file name

            x = filename.split(path + '\\')
            tempList.append(x[1])
            print(x[1])
           
            # Scrape the full contents of the file
           
            fullContentsOfFile = f.read()
            print(fullContentsOfFile)

            fileNameList.append(scrapeCsDataPoints(tempList, fullContentsOfFile))
           
   
    tempList = []
    return fileNameList        
   


def getFileNames():
   
    path = 'C:\\InsertYourFilePath'
    FileNameList1 = openFiles(path)
   
    path = 'C:\\InsertAnotherFilePath'
    FileNameList2 = openFiles(path)
   
    print(FileNameList1)
    print(len(FileNameList1))
    print('\r\n' + '\r\n')
    print(FileNameList2)
    print(len(FileNameList2))
   

   
def scrapeCsDataPoints(tempList, fullContentsOfFile):
   
    #####   Split and cast for each individual data point we want to compare
   
    # Number of Iterations

    # This is the text to the left of the data you want
    left = 'Iterations: ' 

    # This is the text to the right of the data you want 
    right = 'Ending1'   

    # This pulls the text you want and then strips it of white space on both the left and right side if there's any
    iterations = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()

    # Append to the list of variables you're collecting from this file
    tempList.append(iterations)
    print(iterations)
   
    # Final Delta-Sum
    left = 'Delta-Sum: '
    right = 'Ending2'
    deltaSum = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(deltaSum)
    print(deltaSum)
   
    # Initial Sum of Squares Residual
    left = 'Initial sum of squares of residual:'
    right = 'Ending3'
    initialSumOfSquaresResidual = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(initialSumOfSquaresResidual)
    print(initialSumOfSquaresResidual)
   
    # Final Sum of Squares Residual
    left = 'Final sum of squares of residual:'
    right = 'Ending4'
    finalSumOfSquaresResidual = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(finalSumOfSquaresResidual)
    print(finalSumOfSquaresResidual)
   
    # Initial Rotation
    left = 'Initial rotation:'
    right = 'Ending5'
    initialRotation = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(initialRotation)
    print(initialRotation)
   
    # Corrected Rotation
    left = 'Corrected rotation:'
    right = 'Ending6'
    correctedRotation = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(correctedRotation)
    print(correctedRotation)
   
    # Found Pattern Center Offset
    left = 'Found pattern center offset:'
    right = 'Ending7'
    patternCenterOffset = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(patternCenterOffset)
    print(patternCenterOffset)
   
    # Calibrated Magnification
    left = 'Calibrated magnification:'
    right = 'Ending8'
    calibratedMagnification = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(calibratedMagnification)
    print(calibratedMagnification)
   
    # M
    left = 'M:'
    right = 'K:'
    M = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(M)
    print(M)
   
    # K
    left = 'K:'
    right = 'P1:'
    K = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(K)
    print(K)
   
    # P1
    left = 'P1:'
    right = 'P2:'
    P1 = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(P1)
    print(P1)
   
    # P2
    left = 'P2:'
    right = 'P3:'
    P2 = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(P2)
    print(P2)
   
    # P3
    left = 'P3:'
    right = 'Calibrated distortion results in image space (pixels):'
    P3 = fullContentsOfFile[fullContentsOfFile.index(left)+len(left):fullContentsOfFile.index(right)].strip()
    tempList.append(P3)
    print(P3)
       
    return tempList



getFileNames()
```
